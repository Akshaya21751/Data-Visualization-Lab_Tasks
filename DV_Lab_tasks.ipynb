```python
#Task 2b:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Dataset
df = pd.read_csv("/content/sample_data/test.csv").head(50)  # Take first 50 rows for clarity
print("First 5 rows of dataset:")
print(df.head())

# Step 2: Select a Continuous Variable (you can change this)
ram = df['ram']
battery = df['battery_power']
memory = df['int_memory']

# Step 3: Summary Statistics
print("\nRAM Summary Statistics:")
print(ram.describe())

print("\nBattery Power Summary Statistics:")
print(battery.describe())

print("\nInternal Memory Summary Statistics:")
print(memory.describe())

# Step 4: Histogram (Distribution)
plt.figure(figsize=(7,4))
sns.histplot(ram, bins=10, kde=False, color='skyblue', edgecolor='black')
plt.title("Histogram of RAM")
plt.xlabel("RAM (in MB)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Step 5: Box Plot (Outliers and Spread)
plt.figure(figsize=(7,3))
sns.boxplot(x=ram, color='lightgreen')
plt.title("Box Plot of RAM")
plt.xlabel("RAM (in MB)")
plt.tight_layout()
plt.show()

# Step 6: Density Plot (Smooth Distribution)
plt.figure(figsize=(7,4))
sns.kdeplot(ram, shade=True, color='coral')
plt.title("Density Plot of RAM")
plt.xlabel("RAM (in MB)")
plt.tight_layout()
plt.show()

# Step 7: Rug Plot (Exact Value Indicators)
plt.figure(figsize=(7,3))
sns.rugplot(ram, height=0.1, color='purple')
plt.title("Rug Plot of RAM")
plt.xlabel("RAM (in MB)")
plt.tight_layout()
plt.show()

# Step 8: Scatter Plot (Relationship between Battery Power and RAM, colored by Price Range)
x_feature = 'battery_power'
y_feature = 'ram'
target = 'pc' # Assuming 'pc' is the price range column

plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x=x_feature, y=y_feature, hue=target, palette='viridis', s=80)
plt.title("Scatter Plot - Battery Power vs RAM (Colored by Price Range)", fontsize=14)
plt.xlabel("Battery Power (mAh)")
plt.ylabel("RAM (MB)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Step 9: Line Plot (Average RAM by Price Range)
avg_ram = df.groupby(target)['ram'].mean().reset_index()

plt.figure(figsize=(7,5))
sns.lineplot(data=avg_ram, x=target, y='ram', marker='o', color='teal', linewidth=2)
plt.title("Line Plot - Average RAM by Price Range", fontsize=14)
plt.xlabel("Price Range (0 = Low, 3 = High)")
plt.ylabel("Average RAM (MB)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Task 3:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Step 1: Load dataset (only first 15 rows)
df = pd.read_csv("/content/sample_data/test.csv").head(15)
print("First 5 rows of dataset:")
print(df.head())

# Step 2: Convert categorical columns to string (for plotting)
df['three_g'] = df['three_g'].astype(str)
df['four_g'] = df['four_g'].astype(str)
df['wifi'] = df['wifi'].astype(str)
df['touch_screen'] = df['touch_screen'].astype(str)

sns.set(style="whitegrid")

# =====================================================
# 1 Stacked Bar Chart: 3G vs 4G
# =====================================================
plt.figure(figsize=(6,4))
pd.crosstab(df['three_g'], df['four_g']).plot(
    kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'], figsize=(6,4))
plt.title("Stacked Bar Chart: 3G vs 4G")
plt.xlabel("3G (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# =====================================================
# 2 Grouped Bar Chart: 3G vs 4G
# =====================================================
plt.figure(figsize=(6,4))
pd.crosstab(df['three_g'], df['four_g']).plot(
    kind='bar', stacked=False, color=['#1f77b4', '#ff7f0e'], figsize=(6,4))
plt.title("Grouped Bar Chart: 3G vs 4G")
plt.xlabel("3G (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# =====================================================
# 3 Segmented Bar Chart (%): 3G vs 4G
# =====================================================
plt.figure(figsize=(6,4))
(pd.crosstab(df['three_g'], df['four_g'], normalize='index') * 100).plot(
    kind='bar', stacked=True, cmap='viridis', figsize=(6,4))
plt.title("Segmented Bar Chart (%): 3G vs 4G")
plt.xlabel("3G (0 = No, 1 = Yes)")
plt.ylabel("Percentage (%)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# =====================================================
# 4 Mosaic Plot: WiFi vs Touch Screen
# =====================================================
plt.figure(figsize=(6,6))
mosaic(df, ['wifi', 'touch_screen'])
plt.title("Mosaic Plot: WiFi vs Touch Screen")
plt.tight_layout()
plt.show()

# =====================================================
# 5 Scatter Plot with Fit Line: RAM vs Battery Power
# =====================================================
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='ram', y='battery_power', color='skyblue', s=70)
sns.regplot(data=df, x='ram', y='battery_power', scatter=False, color='red')
plt.title("Scatter Plot with Fit Line: RAM vs Battery Power")
plt.xlabel("RAM (MB)")
plt.ylabel("Battery Power (mAh)")
plt.tight_layout()
plt.show()

# =====================================================
# 6 Average Mobile Weight by Touch Screen (Bar Chart)
# =====================================================
plt.figure(figsize=(7,4))
df.groupby('touch_screen')['mobile_wt'].mean().plot(kind='bar', color='lightgreen')
plt.title("Average Mobile Weight by Touch Screen")
plt.xlabel("Touch Screen (0 = No, 1 = Yes)")
plt.ylabel("Average Mobile Weight")
plt.tight_layout()
plt.show()

# =====================================================
# 7 Grouped Kernel Density: Mobile Weight by Touch Screen
# =====================================================
plt.figure(figsize=(7,4))
sns.kdeplot(data=df, x='mobile_wt', hue='touch_screen', fill=True, alpha=0.6, palette='coolwarm')
plt.title("Grouped Kernel Density: Mobile Weight by Touch Screen")
plt.xlabel("Mobile Weight")
plt.tight_layout()
plt.show()

# =====================================================
# 8 Box Plot: Touch Screen vs Mobile Weight
# =====================================================
plt.figure(figsize=(7,4))
sns.boxplot(data=df, x='touch_screen', y='mobile_wt', palette='mako')
plt.title("Box Plot: Touch Screen vs Mobile Weight")
plt.xlabel("Touch Screen (0 = No, 1 = Yes)")
plt.ylabel("Mobile Weight")
plt.tight_layout()
plt.show()
```

```python
#Task 4:
# Step 1  Load the Dataset (only first 15 rows)
import pandas as pd

df = pd.read_csv("/content/sample_data/test.csv").head(15)   # <-- limit to first 15 rows
print(df.head())
print(df.info())

# Step 2  Identify Numeric & Categorical Columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Numeric Columns:", num_cols)
print("Categorical Columns:", cat_cols)

# Step 3  Scatterplot Matrix (pairwise numeric relationships)
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[num_cols[:5]], diag_kind='kde')
plt.suptitle("Scatterplot Matrix (First 15 Rows)", y=1.02)
plt.show()

# Step 4  Parallel Coordinates (compare numeric features across a class)
from pandas.plotting import parallel_coordinates

# If you have a target column (e.g., 'price_range'), use it for coloring
# If not categorical, convert to string to treat as category
if 'price_range' in df.columns:
    target_col = 'price_range'
else:
    # fallback: take last numeric column as a pseudo category
    target_col = num_cols[-1]
    df[target_col] = df[target_col].astype(str)

subset_cols = num_cols[:5] + [target_col]
parallel_coordinates(df[subset_cols], class_column=target_col, colormap='viridis')
plt.title("Parallel Coordinates Plot (First 15 Rows)")
plt.show()

# Step 5  Line Graph (numeric measures over index)
df[num_cols[:4]].plot(kind='line', figsize=(10,5))
plt.title("Line Graph of Multiple Measures (First 15 Rows)")
plt.xlabel("Record Index")
plt.ylabel("Value")
plt.show()

# Step 6  Stacked Bar Chart (categorical vs categorical)
# If dataset has no object columns, create a categorical bin to demonstrate
if len(cat_cols) >= 2:
    stacked_data = pd.crosstab(df[cat_cols[0]], df[cat_cols[1]])
    stacked_data.plot(kind='bar', stacked=True, figsize=(8,5))
    plt.title(f"Stacked Bar Chart: {cat_cols[0]} vs {cat_cols[1]} (First 15 Rows)")
    plt.ylabel("Count")
    plt.show()
else:
    print("Not enough categorical columns for stacked bar chart.")
```

```python
#Task 5b:
import plotly.graph_objects as go
import pandas as pd

# Load dataset
df = pd.read_csv("/content/sample_data/test.csv")

# Let's say we want top 5 phones by RAM in each (four_g, three_g, wifi) group
top_devices = df.groupby(['four_g', 'three_g', 'wifi']).apply(lambda x: x.nlargest(5, 'ram')).reset_index(drop=True)

# Create labels (combine hierarchy + device id)
labels = top_devices['four_g'].astype(str) + 'G-Four | ' + \
         top_devices['three_g'].astype(str) + 'G-Three | ' + \
         'WiFi:' + top_devices['wifi'].astype(str) + ' | ID:' + top_devices['id'].astype(str)

# Create Sunburst Chart
fig = go.Figure(go.Sunburst(
    labels=labels,
    parents=[''] * len(top_devices),   # top-level (simplified)
    values=top_devices['ram']
))

fig.update_layout(title='Top 5 Devices by RAM for each FourG-ThreeG-WiFi Group')
fig.show()
```

```python
#Task 6
# TASK 6  Mobile Price Classification Network (Weighted Force Layout)
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Load the dataset (only first 15 rows for clarity)
df = pd.read_csv("/content/sample_data/test.csv").head(15)

print("Dataset shape:", df.shape)
print(df.head())

# Step 2: Identify target column (price_range)
if "price_range" in df.columns:
    target_col = "price_range"
else:
    target_col = df.columns[-1]  # fallback

# Step 3: Create a graph
G = nx.Graph()

# Select a few features for readability (e.g., first 6 columns excluding target)
features = df.columns.drop(target_col).tolist()[:6]

# Add feature nodes and class (price_range) nodes
G.add_nodes_from(features, type="feature")
G.add_nodes_from(df[target_col].unique(), type="class")

# Step 4: Add weighted edges (feature  price_range)
for _, row in df.iterrows():
    class_label = row[target_col]
    for feature in features:
        if G.has_edge(feature, class_label):
            G[feature][class_label]['weight'] += 1
        else:
            G.add_edge(feature, class_label, weight=1)

# Step 5: Layout and coloring
pos = nx.spring_layout(G, seed=42, k=0.5)
color_map = []
for node, attr in G.nodes(data=True):
    if attr["type"] == "feature":
        color_map.append("skyblue")     # Features = blue
    else:
        color_map.append("lightcoral")  # Classes = red

# Step 6: Draw the network graph
plt.figure(figsize=(12, 8))
edges = G.edges(data=True)

# Draw nodes and edges
nx.draw(
    G, pos,
    with_labels=True,
    node_size=1500,
    node_color=color_map,
    font_size=10,
    font_weight="bold",
    width=[d['weight'] for (_, _, d) in edges],
    edge_color="gray"
)

# Add edge weight labels
edge_labels = {(u, v): d['weight'] for u, v, d in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

plt.title("Mobile Price Classification Network (first 15 rows, weighted with labels)")
plt.show()
```

```python
#Task 7:
# Word Cloud for Mobile Price Classification Dataset
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("/content/sample_data/test.csv").head(50)  # use first 50 rows for clarity

# Step 2: Combine feature names and values into a text corpus
# Example: "battery 1500 ram 512 memory 8 range 1"
text_data = []
for _, row in df.iterrows():
    sentence = f"battery {row['battery_power']} ram {row['ram']} memory {row['int_memory']}"
    text_data.append(sentence)

# Join all rows into a single text string
text = " ".join(text_data)

# Step 3: Generate Word Cloud
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      colormap='viridis',
                      collocations=False).generate(text)

# Step 4: Display the Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud - Mobile Price Classification Dataset", fontsize=14)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


# Load dataset
data = pd.read_csv('/content/sample_data/test.csv').head(50)


# Select feature (RAM)
feature = 'ram'
x = np.arange(len(data)).reshape(-1, 1)
y = data[feature].values


# Handle missing values
y = np.nan_to_num(y, nan=np.nanmean(y))


# Linear Regression
model = LinearRegression()
model.fit(x, y)
slope = model.coef_[0]
intercept = model.intercept_


# Plot feature with trend line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='skyblue', label='RAM Data')
plt.plot(x, slope * x + intercept, color='red', linewidth=2, label='Trend Line')
plt.title('Trend Line of RAM Across Mobile Samples')
plt.xlabel('Record Index')
plt.ylabel('RAM Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
