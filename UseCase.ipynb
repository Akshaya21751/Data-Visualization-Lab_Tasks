```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
df.groupby('Region')['Total_Revenue'].sum().plot(kind='bar')
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.show()
```

Line Chart


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
df.groupby('Ship_Date')['Total_Revenue'].sum().plot(kind='line')
plt.title('Revenue Over Time')
plt.xlabel('Ship Date')
plt.ylabel('Total Revenue')
plt.show()
```

Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
sns.heatmap(df[['Unit_Cost','Total_Revenue','Total_Profit']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(6,6))
df.groupby('Item_Type')['Total_Revenue'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Revenue Share by Item Type')
plt.ylabel('')
plt.show()
```

Box Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
sns.boxplot(x='Sales_Channel', y='Total_Profit', data=df)
plt.title('Profit Distribution by Sales Channel')
plt.show()
```

Scatter Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
sns.scatterplot(x='Unit_Cost', y='Total_Revenue', hue='Sales_Channel', data=df)
plt.title('Unit Cost vs Total Revenue')
plt.show()
```

Leaderboard Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
leaderboard = df.groupby('Country')['Total_Revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=leaderboard.values, y=leaderboard.index, palette='viridis')
plt.title('Top 10 Countries by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Country')
plt.show()
```

Area Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
df.groupby('Ship_Date')['Total_Profit'].sum().plot(kind='area', alpha=0.5)
plt.title('Profit Over Time')
plt.xlabel('Ship Date')
plt.ylabel('Total Profit')
plt.show()
```

Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
plt.figure(figsize=(8,5))
plt.hist(df['Total_Revenue'], bins=20, edgecolor='black')
plt.title('Revenue Distribution')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.show()
```

Funnel Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("100_Sales.csv").dropna(axis=1, how='all')
funnel_data = df.groupby('Item_Type')['Total_Revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
plt.barh(funnel_data.index, funnel_data.values, color='skyblue')
plt.gca().invert_yaxis()
plt.title('Revenue Funnel by Item Type')
plt.xlabel('Total Revenue')
plt.ylabel('Item Type')
plt.show()
```
